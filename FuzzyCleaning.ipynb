```python
import numpy as np
from fuzzywuzzy import fuzz
import pandas
cluster_list = []
```

```python
df1 = pandas.read_csv("2438_3279.csv")
df1
```

Strip out [unclear] flags to stop them all clustering based on that string. 

```python
for row in df1.index:
    try:
        df1.iloc[row,8] = df1.iloc[row,8].replace("[unclear]", "")
        df1.iloc[row,8] = df1.iloc[row,8].replace("[/unclear]", "")
        df1.iloc[row,9] = df1.iloc[row,9].replace("[unclear]", "")
        df1.iloc[row,9] = df1.iloc[row,9].replace("[/unclear]", "")
    except AttributeError:
        continue
df1
```

Extract the collector columns and get rid of any duplicates:

```python
name_set = set(df1["Collector1"])
name_set |= set(df1["Collector2"])
name_list = list(name_set)
i = 0
for name in name_list:
    i+= 1

print("Size of unique name list: " + str(i))
```

The function below takes the name_list we created in the last cell and clusters 'similar' names together in a new 2D list:

```python
def token_set_clustering(input_list):
    checked = []

    for name in input_list:
        cluster = []

        # Create a new list of names we haven't encountered so far + turn into a list
        unchecked_list = np.setdiff1d(input_list, checked, assume_unique=True).tolist()

        # Go through each 'new' name and check for similarity against current name
        for unchecked_name in unchecked_list:
            # Only consider if similar enough and new to this iteration
            if (fuzz.token_set_ratio(name, unchecked_name) >= 90) and (unchecked_name not in checked):
                checked.append(unchecked_name)
                cluster.append(unchecked_name)

        if len(cluster) != 0:
            cluster_list.append(cluster)
 
```

```python
token_set_clustering(name_list)
print("Number of names after clustering: " + str(len(cluster_list)))
cluster_list.sort()

for name_list in cluster_list:
    print(name_list)
```

Create a dictionary using the 'best names' (here arbitrarily using the first name listed in each cluster for purposes of illustration + testing). Key should be incorrect name string, value is normalised name string. 

```python
name_dict = {}

for name_list in cluster_list:
    if len(name_list) > 1:
        for name in name_list:
            name_dict[name] = name_list[0]

for key, value in name_dict.items():
    print(key + ": " + value)
```

```python
for row in df1.index:
    try:
        df1.iloc[row,8] = name_dict.get(df1.iloc[row,8], df1.iloc[row,8])
        df1.iloc[row,9] = name_dict.get(df1.iloc[row,9], df1.iloc[row,9])
    except AttributeError:
        continue
df1
```
